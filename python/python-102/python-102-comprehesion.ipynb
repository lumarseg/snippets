### List Comprehension

List comprehesion es una forma de generar lista con una sintaxis más corta.

```python
# List generation by operators.

list_numbers = []
for element in range(1, 11):
    list_numbers.append(element)
print(list_numbers)
```

List Comprehesion nos permite reducir el numero de líneas para la creación de la lista de número, ejemplo:

```python
# List generation by list comprehension.

list_numbers = []
list_numbers = [element for element in range(1, 11)]
print(list_numbers)
```

Se puede incluir operadores aritméticos.

```python
list_numbers = []
list_numbers = [element * 2 for element in range(1, 11)]
print(list_numbers)
```

En el siguiente ejemplo, se puede agregar condiciones al iterable.

```python
# List generation by operators and conditionals

list_numbers = []
for i in range(1, 11):
    if i % 2 == 0:
        list_numbers.append(i)
print(list_numbers)
```

```python
# List generation by list comprehension and conditionals

list_numbers = []
list_numbers = [i for i in range(1, 11) if i % 2 == 0]
print(list_numbers)
```

### Dictionary Comprehension

Se generará un diccionario, las llaves de cada elemento será los valores iterables de 1 hasta menor a 5. El valor de cada elemento será el valor de la iteración multiplicado por 2.

```python
# Dictionary generation by operators

dict = {}
for i in range(1, 5):
    dict[i] = i * 2
print(dict)
```

```python
# Dictionary generation by dictionary comprehension

dict = {}
dict = { i: i * 2 for i in range(1,5)}
print(dict)
```

El siguiente código consiste de la Generación de un Diccionario a partir de una Lista.

```python
countries = ["col", "mex", "bol", "pe"]
print(countries)
```

```python
import random

# Dictionary generation by operators and a List Source.
population = {}
for country in countries:
    population[country] = random.randint(1, 100)
print(population)

# Dictionary generation by dictionary comprehension and a List Source
population = {}
population = {country: random.randint(1, 100) for country in countries}
print(population)
```

El siguiente ejemplo consiste en iterar dos listas en un diccionario.

```python
names = ["Beto", "Zule", "Santi"]
ages = [12, 56, 98]

# Print a reference
print(zip(names, ages))

# Print a List with tuples.
print(list(zip(names, ages)))

# Dictionary generation by dictionary comprehension and a List Source
new_dict = {name: age for (name, age) in zip(names, ages)}
print(new_dict)
```

En este ejemplo, continuamos con Dictionary Comprehention pero con condicionales.

```python
# List of countries
countries = ["col", "mex", "bol", "pe"]

# Dictionary of Population
population = {country: random.randint(1, 100) for country in countries}
print(population)

# Dictionary of Population with conditional
result = {country: population for (country, population) in population.items() if population > 20}
print(result)
```
