## CSV

Pandas tiene la capacidad multiples formatos de archivos. En el caso de CSV, se el simblo del separador fuera otro caracter, se puede indicar como argumento detro del método "read_csv".

```python
pd.read_csv("data/bestsellers-with-categories.csv", sep=",")
```

Si lo se dispone de un header, se puede definir el "Header" en los argumentos. En el siguiente ejemplo, definimos la fila 2 como la fila de encabezado.

```python
pd.read_csv("data/bestsellers-with-categories.csv", sep=",", header=2)
```

```python
df_books = pd.read_csv("data/bestsellers-with-categories.csv")
```
