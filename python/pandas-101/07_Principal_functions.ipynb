# Funciones principales en PANDAS

Algunas de las funciones principales de Pandas son:

1. **read_csv**: esta función permite leer un archivo CSV y cargar los datos en un DataFrame de Pandas.
2. **head**: esta función muestra las primeras filas de un DataFrame.
3. **tail**: esta función muestra las últimas filas de un DataFrame.
4. **describe**: esta función muestra un resumen estadístico del DataFrame, incluyendo el conteo de filas, la media, la desviación estándar, los cuartiles, etc.
5. **info**: esta función muestra información sobre el DataFrame, como el tipo de cada columna y el número de no-nulos.
6. **columns**: esta propiedad muestra el nombre de las columnas del DataFrame.
7. **dtypes**: esta propiedad muestra el tipo de datos de cada columna del DataFrame.
8. **shape**: esta propiedad muestra el tamaño del DataFrame en forma de tupla (filas, columnas).
9. **sort_values**: esta función permite ordenar el DataFrame por una o varias columnas.
10. **groupby**: esta función permite agrupar el DataFrame por una o varias columnas y aplicar una función a cada grupo.
