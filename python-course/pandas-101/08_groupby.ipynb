# Groupby

```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
df_books = pd.read_csv("data/bestsellers-with-categories.csv", sep=",", header=0)
df_books.head(2)
```

```python
df_books.groupby("Author").count()
```

```python
df_books.groupby("Author").sum().loc["William Davis"]
```

```python
df_books.groupby("Author").mean().reset_index().sort_values("User Rating", ascending=False).drop(["Year","Reviews","Price"], axis=1)
```

Funciones de agregación

```python
df_books.groupby("Author").agg(["min", "max"]).drop(["Name", "Genre", "Reviews"],axis=1)
```

```python
df_books.groupby("Author").agg({"User Rating": ["min", "max"], "Reviews": "sum"})
```

```python
df_books.groupby(["Author", "Year"]).count()
```
